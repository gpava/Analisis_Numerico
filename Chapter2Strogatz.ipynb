```python
def graph(f,x):
  x_dot=f
  fig = plt.figure()
  ax = fig.add_subplot(1,1,1)
  ax.spines['left'].set_position(('data',0.0))
  ax.spines['bottom'].set_position(('data',0.0))
  ax.spines['top'].set_color('none')
  ax.spines['right'].set_color('none')
  plt.plot(x,x_dot)
  plt.grid()
```

```python
def step_euler(y,h,f):
  next_y=y+h*f(y)
  return next_y

def euler (h,f,y_0,start_t,end_t):
  N= int((end_t - start_t)/h)
  t_list=np.linspace(start_t,end_t,N+1)

  y_list=np.zeros(N+1)

  y_list[0]=y_0

  for i in range(0,N):
    y_list[i+1]=step_euler(y_list[i],h,f)

  return t_list,y_list
```

###Ejercicio 2.2.1
\begin{equation}
\frac{dx}{dt}=4x^2-16
\end{equation}


###Los puntos fijos son $x_{1}^{*} = 2$ y $x_{2}^{*} = -2$





```python
x=np.linspace(-5,5,100)
f=4*x**2 - 16
graph(f,x)
```

```python
y_0=0.1
h=0.01
t_0=0
t_N=10

for i in np.arange(-2,4,0.1):
  t_list,y_list=euler(h,fun,i,t_0,t_N)
  plt.plot(t_list,y_list)
  plt.axis([0,0.75,-2.5,6])
  #print(np.abs(y_list[-1]-np.power(2,t_list[-1])))
  plt.grid()
```

### Campos de Pendientes de Ecuaciones Diferenciales

```python
def slopefield(x,y,h):
  # para x,y
  for j in x:
      for k in y:
          slope = g(j,k)
          domain = np.linspace(j-0.07,j+0.07,2)
          def function(x1,y1):
              z = slope*(domain-x1)+y1
              return z
          plt.plot(domain,function(j,k),solid_capstyle='projecting',solid_joinstyle='bevel')
  plt.grid(True)
  plt.show()
```
