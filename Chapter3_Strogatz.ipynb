```

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.optimize import fsolve
import warnings

warnings.filterwarnings('ignore')
```

```
def graph2(f,x,r):
  x_dot=f
  fig = plt.figure()
  ax = fig.add_subplot(1,1,1)
  ax.spines['left'].set_position(('data',0.0))
  ax.spines['bottom'].set_position(('data',0.0))
  ax.spines['top'].set_color('none')
  ax.spines['right'].set_color('none')
  plt.plot(x,x_dot)
  plt.grid()

  x=sp.symbols('x')
  x_dot1=function(x,r)
  roots=sp.solve(x_dot1)
  print(roots)
  for j in range(len(roots)):
    plt.plot(roots[j],0,'ro')
```

```
def graph3(f,x,r):
  x_dot=f
  fig = plt.figure()
  ax = fig.add_subplot(1,1,1)
  ax.spines['left'].set_position(('data',0.0))
  ax.spines['bottom'].set_position(('data',0.0))
  ax.spines['top'].set_color('none')
  ax.spines['right'].set_color('none')
  plt.plot(x,x_dot)
  plt.grid()
```

```
def function(x,r):
  """Ejercicio 3.1.1"""
  return 1+r*x+x**2
  #return r+(1/2)*x - x/(x+1)#
```

### Ejercicio 3.1.1
\begin{equation}
\dot { x } =1+rx+x^2
\end{equation}

###Con $r=2$

```
x=np.linspace(-10,10,100)
r=2
f=1+r*x+x**2
graph2(f,x,r)
```

```
r=2
b=2
ti=0; tf=1; dt=100
tspam=[ti, tf]
t=np.linspace(ti,tf,dt)
for x0 in np.arange(-5,7,0.5):
  sol=solve_ivp(fun,tspam,[x0],args=(r,b),t_eval=t)
  plt.plot(sol.t, sol.y[0])

plt.axis([ti, tf, -10, 10])
plt.grid()
```

### Ejercicio 3.1.2
\begin{equation}
\dot { x } = r -cosh(x)
\end{equation}

### Con $r\geq 1$

```
x=np.linspace(-3,3,100)
#r=1#
r=2
f = r-np.cosh(x)
graph3(f,x,r)

def roots(x):
  return r-np.cosh(x)

root = fsolve(roots,[-2,2])

for j in range(len(root)):
    plt.plot(root[j],0,'ro')
```

```
#Gráfica de los flujos
r=2

def fun (t,x,r):
  return r-np.cosh(x)

ti=0; tf=6; dt=1000
tspam=[ti, tf]
t=np.linspace(ti,tf,dt)
for x0 in np.arange(0,5,0.2):
  sol=solve_ivp(fun,tspam,[x0],args=(r,),t_eval=t)
  plt.plot(sol.t, sol.y[0])

plt.axis([ti, tf, 0, 4])
plt.grid()
```

```
#Gráfica del diagrama de bifurcaciones

ti=0; tf=10; dt=100
tspam=[ti, tf]

for r in np.arange(1,3,0.1):
  for x0 in np.arange(0,5,0.1):
    sol=solve_ivp(fun,tspam,[x0],args=(r,))
    plt.scatter(r, sol.y[0][-1], color='darkgreen', marker='.')

tspam=[ti, -tf]

for r in np.arange(1,3,0.1):
  for x0 in np.arange(0,5,0.1):
    sol=solve_ivp(fun,tspam,[x0],args=(r,))
    plt.scatter(r, sol.y[0][-1], color='red', marker='.')


plt.axis([0, 3, -2.5, 2.5])
plt.grid()
```

### Ejercicio 3.1.3
\begin{equation}
\dot { x } = r+x-ln(1+x)
\end{equation}

```
x=np.linspace(-10,10,100)
r=-1
f = r + x - np.log(1+x)
graph3(f,x,r)

def roots(x):
  return r +x - np.log(1+x)

root = fsolve(roots,[-0.8,7])

for j in range(len(root)):
  plt.plot(root[j],0,'ro')
```

```
#Gráfica de los flujos

def fun3(t,x,r):
  return r+x-np.log(1+x)


r=-1

ti=0; tf=4; dt=100
tspam=[ti, tf]
t=np.linspace(ti,tf,dt)
for x0 in np.arange(0,7,0.3):
  sol=solve_ivp(fun3,tspam,[x0],args=(r,),t_eval=t)
  plt.plot(sol.t, sol.y[0])

plt.axis([ti, tf, 0, 150])
plt.grid()
```

```
#Gráfica del diagrama de bifurcaciones

ti=0; tf=10; dt=100
tspam=[ti, tf]

for r in np.arange(-1,0,0.1):
  for x0 in np.arange(0,5,0.1):
    sol=solve_ivp(fun3,tspam,[x0],args=(r,))
    plt.scatter(r, sol.y[0][-1], color='darkgreen', marker='.')

tspam=[ti, -tf]

for r in np.arange(-1,0,0.1):
  for x0 in np.arange(0,5,0.1):
    sol=solve_ivp(fun3,tspam,[x0],args=(r,))
    plt.scatter(r, sol.y[0][-1], color='red', marker='.')


plt.axis([-1.3, 0.5, -2.5, 2.5])
plt.grid()
```
